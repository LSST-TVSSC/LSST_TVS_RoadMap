```python
import numpy as np
import matplotlib.pyplot as plt
get_ipython().magic('matplotlib inline')
```

```python
# template from Peter Nugent 
# import and plot first template
ascifile = 'sn91bg_lc.v1.1.dat'
data =np.genfromtxt(ascifile)
phase,U,B,V,R,I = zip(*data)
plt.plot(phase,U,'-r', label='U')
plt.plot(phase,B,'-b',label='B')
plt.plot(phase,V,'-g',label='V')
plt.plot(phase,R,'-c',label='R')
plt.plot(phase,I,'-m',label='I')
plt.ylim(5,-1)
plt.title('1991bg')
plt.legend(loc=1)
```

```python
# template from Peter Nugent 
ascifile = 'sn1a_lc.v1.2.dat'
data =np.genfromtxt(ascifile)
phase,U,B,V,R,I,J,H,K = zip(*data)
plt.plot(phase,U,'-r', label='U')
plt.plot(phase,B,'-b',label='B')
plt.plot(phase,V,'-g',label='V')
plt.plot(phase,R,'-c',label='R')
plt.plot(phase,I,'-m',label='I')
plt.plot(phase,J,'-k',label='J')
plt.plot(phase,H,'-',color='orange',label='H')
plt.plot(phase,K,'-y',label='K')
plt.ylim(5,-1)
plt.title('SN Ia normal')
plt.legend(loc=1)
```

```python
# Template 2013by  (Valenti et al 2015)
ascifile = 'IIL.txt'
data =np.genfromtxt(ascifile)
phase,u,B,V,r,i = zip(*data)
plt.plot(phase,u-min(B),'-r', label='u')
plt.plot(phase,B-min(B),'-b',label='B')
plt.plot(phase,V-min(B),'-g',label='V')
plt.plot(phase,r-min(B),'-c',label='r')
plt.plot(phase,i-min(B),'-m',label='i')
plt.ylim(7,-1)
plt.xlim(-2,150)
plt.title('IIL (2013by)')
plt.legend(loc=1)
```

```python
# Template from  Valenti et al 2011
ascifile = 'Ib.txt'
data =np.genfromtxt(ascifile)
phase,u,B,V,r,i = zip(*data)
plt.plot(phase,u-min(B),'-r', label='u')
plt.plot(phase,B-min(B),'-b',label='B')
plt.plot(phase,V-min(B),'-g',label='V')
plt.plot(phase,r-min(B),'-c',label='r')
plt.plot(phase,i-min(B),'-m',label='i')

# Template from Fremling et al 2014
ascifile = 'Ib2.txt'
data =np.genfromtxt(ascifile)
phase,u,B,V,r,i = zip(*data)
plt.plot(phase,u-min(B),'--r', label='u')
plt.plot(phase,B-min(B),'--b',label='B')
plt.plot(phase,V-min(B),'--g',label='V')
plt.plot(phase,r-min(B),'--c',label='r')
plt.plot(phase,i-min(B),'--m',label='i')

plt.ylim(5,-1)
plt.xlim(-2,150)
plt.title('Ib (2009jf,ptf13bvn)')
plt.legend(loc=1)
```
